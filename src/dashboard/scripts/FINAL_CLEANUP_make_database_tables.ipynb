```python
import pandas as pd
import random
import string
from faker import Faker

# Set the random seed for reproducibility
random.seed(42)

# Initialize the Faker generator
fake = Faker()

# Define the length of the DataFrame
length = 71

# Generate random topic IDs
topic_ids = list(range(1, length+1))

# Generate random topic names
topic_names = [fake.word().capitalize() for _ in range(length)]
# Generate random descriptions using lorem ipsum words
topic_group = [fake.word().capitalize() for _ in range(length)]

descriptions = [fake.paragraph(nb_sentences=4) for _ in range(length)]

# Generate random narratives using lorem ipsum words
narratives = [fake.sentence(nb_words=20) for _ in range(length)]

# Create the DataFrame
df = pd.DataFrame({
    'topic_id': topic_ids,
    'topic_name': topic_names,
    'topic_group': topic_group,
    'description': descriptions,
    'narrative': narratives
})

# Print the DataFrame
df.to_csv(BASE.joinpath('TOPICS_TABLE.csv'), index=False)
display(df)
```

```python
ics = pd.read_csv(BASE.joinpath('ICS_DATABASE_TABLE.csv'))
ics = ics[[x for x in ics.columns if not x in ["institution_name", "countries", "funders"]]]
ics.to_csv('ICS_DATABASE_TABLE.csv', index=False)
```

```python
uoa_ids = sorted(ics.uoa.unique())
uoa_names = [fake.word().capitalize() for _ in range(len(ics.uoa.unique()))]
df_uoas = pd.DataFrame({
    'uoa_id': uoa_ids,
    "name": uoa_names
})
df_uoas.to_csv(BASE.joinpath('UOA_TABLE.csv'), index=False)
df_uoas
```
